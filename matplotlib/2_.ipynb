###  Figure
The whole figure. The Figure keeps track of all the child Axes, a group of 'special' Artists (titles, figure legends, colorbars, etc.), and even nested subfigures.


```python
fig = plt.figure()             # an empty figure with no Axes
fig, ax = plt.subplots()       # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
# a figure with one Axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```

```python
import matplotlib.pyplot as plt

# Create the subplot mosaic
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])

# Plot on the different axes
axs['left'].plot([0, 1, 2], [0, 1, 4], label='Left Plot')
axs['right_top'].scatter([0, 1, 2], [0, 1, 4], color='r', label='Right Top Plot')
axs['right_bottom'].bar([0, 1, 2], [3, 2, 5], label='Right Bottom Plot')

# Set titles
axs['left'].set_title('Left Plot')
axs['right_top'].set_title('Right Top Plot')
axs['right_bottom'].set_title('Right Bottom Plot')

# Display the plot
plt.tight_layout()
plt.show()
```

#### Subplot Names:

'left', 'right_top', and 'right_bottom' are labels for the subplot axes that you can reference later.
