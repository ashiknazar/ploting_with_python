### lineplot

```python
import matplotlib.pyplot as plt

# Data for the plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y)

# Adding a title and labels
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

The line fig, ax = plt.subplots() in the second code snippet is part of the matplotlib plotting system and is used to create a figure and an axes object for plotting.

- 1.plt.subplots():

- This is a function from matplotlib.pyplot that creates a figure and one or more axes (i.e., plotting areas) at the same time.
- fig (Figure object): This represents the entire figure or the overall window that will contain the plot. It can contain multiple subplots.
- ax (Axes object): This is the actual plot area (or axes) within the figure where data will be plotted. You can have multiple ax objects in a single fig (if you want multiple subplots), but in this case, it's a single axes for a simple plot.

- 2.The use of `fig,ax`:

- `fig` refers to the whole figure, including all axes, titles, labels, and other elements that are part of the plot.
- `ax ` is the actual area where the graph is drawn, and it is where you will plot the data using functions like `ax.plot()`.

### Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Scatter plot
plt.scatter(x, y)

plt.title("Simple Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

### bar chart

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

# Bar chart
plt.bar(categories, values)

plt.title("Simple Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```

### Histogram

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 6, 7]

# Histogram
plt.hist(data, bins=6)

plt.title("Simple Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

### 4. Customizing Plots

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Customized plot
plt.plot(x, y, color='green', linestyle='--', marker='o')

plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Adding grid
plt.grid(True)

plt.show()
```

### 5. Subplots

```python
import matplotlib.pyplot as plt

# Create a figure with 2 rows and 2 columns of subplots
plt.subplot(2, 2, 1)  # (rows, columns, position)
plt.plot([1, 2, 3], [1, 2, 3])

plt.subplot(2, 2, 2)
plt.bar([1, 2, 3], [10, 20, 30])

plt.subplot(2, 2, 3)
plt.scatter([1, 2, 3], [1, 4, 9])

plt.subplot(2, 2, 4)
plt.hist([1, 2, 3, 4, 5, 6, 7], bins=5)

# Show all the plots
plt.tight_layout()
plt.show()
```

### 6. Saving Plots

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)

# Save the plot as a PNG file
plt.savefig('line_plot.png')

# Display the plot
plt.show()
```
