- In Matplotlib, artists are objects that are used to represent the elements in a plot. These include things like lines, text, axes, titles, legends, grids, and even entire figures. Essentially, anything that is visible in a plot is an artist in Matplotlib. The most common artists are:

- 1. Figure – The entire window or image that holds everything.

- 2. Axes – The coordinate system of the plot (i.e., the area where the data is plotted).

- 3. Axis – The individual X and Y axes.

- 4. Line – A plot line (such as in a line plot).

- 5. Text – Text annotations in the plot.

- 6. Patch – A collection of objects like rectangles, circles, polygons, etc.

- 7. Legend – A box describing the different elements of the plot.


```python
import matplotlib.pyplot as plt

# Create a new figure and axis
fig, ax = plt.subplots()

# Plotting a simple line (Line Artist)
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
ax.plot(x, y, label='y = x^2')  # This is a Line artist

# Adding text (Text Artist)
ax.text(1, 10, 'Parabola', fontsize=12, color='red')  # This is a Text artist

# Adding a title (Text Artist)
ax.set_title('Simple Plot with Artists')

# Adding a grid (Grid artist)
ax.grid(True)  # Grid is a type of artist in Matplotlib

# Adding a legend (Legend Artist)
ax.legend()  # Legend is also an artist

# Display the plot
plt.show()
```

# 1. Figure-related functions

- `plt.figure()`: Create a new figure
   - parameters : 
        - `figsize`: Tuple specifying the figure size `(width, height)`.
        - `dpi`: Dots per inch (resolution of the figure).
        - `facecolor`: Background color of the figure.

# 2. Axes-related functions

- `plt.subplots()`: Creates a figure and a set of subplots (axes).
    - Parameters:
       - `nrows`, `ncols`: Number of rows and columns of subplots.
       - `sharex`, `sharey`: Whether the x or y axes should be shared across subplots.
       - `figsize`: Size of the figure.




## sharex

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and two subplots, sharing the x-axis
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, sharey=False)

# Plot data on both axes
ax1.plot(x, y1, label='sin(x)')
ax1.set_title('Sine Wave')
ax2.plot(x, y2, label='cos(x)', color='r')
ax2.set_title('Cosine Wave')

# Set labels
ax2.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

# Show the plot
plt.tight_layout()
plt.show()
```

- `ax.set_xlabel()`: Sets the label for the x-axis.
   - Parameters:
       - xlabel: The label text.
       - `fontsize`: Font size of the label.
       - `color`: Color of the label.
       - `labelpad`: Distance between label and axis.
       - `loc`: Position of the label (e.g., 'left', 'center', 'right').

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y, label='Sine Wave')

# Set the x-axis label with additional formatting
ax.set_xlabel('Time (seconds)', fontsize=14, color='blue', labelpad=15, loc='right')

# Set the y-axis label
ax.set_ylabel('Amplitude', fontsize=14)

# Set a title
ax.set_title('Sine Wave Example')

# Display a legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

- `ax.set_ylabel()`: Sets the label for the y-axis.
  - Parameters:
    - ylabel: The label text.
    - `fontsize`, `color`, `labelpad`, `loc` (same as set_xlabel).
- `ax.set_title()`: Sets the title of the plot.
  - Parameters:
    - title: The title text.
    - fontsize, color, loc: Font size, color, and position (e.g., 'left', 'center', 'right').
- `ax.set_xlim()`: Sets the limits for the x-axis.

  - Parameters:
    - left, right: The lower and upper bounds of the x-axis.
- `ax.set_ylim()`: Sets the limits for the y-axis.

  - Parameters:
    - bottom, top: The lower and upper bounds of the y-axis.

# 3. Ticks-related functions

- `ax.set_xticks()`: Sets the positions of the ticks on the x-axis.
  - Parameters:
    - ticks: List of tick positions.
- `ax.set_yticks()`: Sets the positions of the ticks on the y-axis.
  - Parameters:
    - ticks: List of tick positions.
- `ax.set_xticklabels()`: Sets the labels for the x-ticks.
  - Parameters:
    - labels: List of labels for the x-ticks.
    - rotation: Angle of rotation for the labels.
- `ax.set_yticklabels()`: Sets the labels for the y-ticks.
  - Parameters:
    - labels: List of labels for the y-ticks.
    - rotation: Angle of rotation for the labels.
- `ax.tick_params()`: Customizes tick parameters.

  - Parameters:
    - axis: 'x', 'y', or 'both' for which axis to modify.
    - direction: 'in', 'out', 'inout' to specify the direction of the ticks.
    - length: Length of the ticks.
    - width: Width of the ticks.
    - colors: Color of the ticks.
    - labelsize: Size of the tick labels.
    - labelrotation: Rotation of tick labels.


```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y, label='Sine Wave')

# Set custom tick positions
ax.set_xticks([0, 2, 4, 6, 8, 10])
ax.set_yticks([-1, -0.5, 0, 0.5, 1])

# Set custom labels for x-ticks and y-ticks
ax.set_xticklabels(['Start', 'Two', 'Four', 'Six', 'Eight', 'End'], rotation=45)
ax.set_yticklabels(['Low', 'Mid-Low', 'Zero', 'Mid-High', 'High'], rotation=90)

# Customize tick parameters
ax.tick_params(axis='x', direction='inout', length=10, width=2, colors='red', labelsize=12, labelrotation=45)
ax.tick_params(axis='y', direction='in', length=6, width=2, colors='blue', labelsize=12, labelrotation=90)

# Set the title, x-label, and y-label
ax.set_title('Sine Wave with Custom Ticks and Labels')
ax.set_xlabel('Time (seconds)')
ax.set_ylabel('Amplitude')

# Display a legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

# 4. Grid-related functions

- `ax.grid()`: Toggles the grid on or off.
   - Parameters:
     - True or False: Turns the grid on or off.
     - which: 'major', 'minor', or 'both' to specify which gridlines to display.
     - axis: 'both', 'x', or 'y' for which axis to display the grid.
     - color: Color of the gridlines.
     - linewidth: Width of the gridlines.
     -   linestyle: Style of the gridlines (e.g., '--', '-.', ':').
     - ax.minorticks_on(): Enables minor ticks.

     - ax.minorticks_off(): Disables minor ticks.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and axis
fig, ax = plt.subplots()

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the data
ax.plot(x, y, label='sin(x)', color='blue')

# Enable minor ticks
ax.minorticks_on()

# Enable the major and minor gridlines
ax.grid(True, which='major', axis='both', color='black', linewidth=0.5, linestyle='-')  # Major gridlines
ax.grid(True, which='minor', axis='both', color='gray', linewidth=0.5, linestyle='--')  # Minor gridlines

# Disable minor ticks (optional)
# ax.minorticks_off()

# Set labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Grid and Ticks Customization')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

# 5. Legend-related functions

- `ax.legend()`: Adds a legend to the plot.
   - Parameters:
      - loc: Location of the legend (e.g., 'best', 'upper right', 'center').
      - fontsize: Font size of the legend.
      - title: Title of the legend.
      - frameon: Whether the legend should have a background box (True or False).

# 6. Plotting-related functions

- ax.plot(): Creates a line plot.
  - Parameters:
    - color: Color of the line.
    - linewidth: Width of the line.
    - linestyle: Style of the line (e.g., '-', '--', '-.', ':').
    - marker: Marker style (e.g., 'o', 'x', '+', '*').
    - label: Label for the plot, used in the legend.
- ax.scatter(): Creates a scatter plot.
  - Parameters:
    - color: Color of the points.
    - s: Size of the points.
    - marker: Shape of the markers.
    - label: Label for the points.
    - ax.bar(): Creates a bar plot.

  - Parameters:
    - color: Color of the bars.
    - width: Width of the bars.
    - label: Label for the bars.
- ax.hist(): Creates a histogram.
  - Parameters:
    - bins: Number of bins.
    - density: Whether to normalize the histogram.
    - color: Color of the bars.
- ax.fill_between(): Fills the area between two horizontal curves.

  - Parameters:
    - y1, y2: Data for the y-values.
    - color: Fill color.
    - alpha: Transparency of the fill.


# 7. Text-related functions

- `ax.text()`: Adds text to the plot at a specific position.
  - Parameters:
    - x, y: Position of the text.
    - s: The text string.
    - fontsize, color: Font size and color of the text.
    - ha, va: Horizontal and vertical alignment of the text.

# 8. Other miscellaneous functions

- ax.set_aspect(): Sets the aspect ratio of the plot.
  - Parameters:
    - 'equal': Ensures equal scaling of x and y axes.
    - 'auto': Let Matplotlib adjust the aspect ratio automatically.
    - Numeric values: Aspect ratio value (e.g., 1.5).
- ax.set_facecolor(): Sets the background color of the axes.
  - Parameters:
    - color: The desired background color.
- ax.set_edgecolor(): Sets the edge color of the axes.
  - Parameters:
    - color: The desired edge color.
- fig.savefig(): Saves the figure to a file.
  - Parameters:
    - filename: Name of the file to save the plot.
    - dpi: Resolution of the saved figure.
    - transparent: Whether to make the background transparent.
