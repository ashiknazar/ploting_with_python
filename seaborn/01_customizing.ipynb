```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
```

#### customizing
- Changing Themes: Seaborn allows you to control the style and color palette of your plots with the set_style() and set_palette() functions.

```python
sns.set_style("darkgrid")
sns.set_palette("Blues")

# Now plot
sns.histplot(data['total_bill'], kde=True)
plt.show()
```

- Customizing Axes Labels and Titles:



```python
sns.scatterplot(data=data, x='total_bill', y='tip')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

- Categorical Plots
   - Seaborn makes it easy to create plots for categorical data.

```python
sns.countplot(data=data, x='day')
plt.show()
```

```python
sns.violinplot(data=data, x='day', y='total_bill')
plt.show()
```

#### Regression Plots
Seaborn can also plot regression lines with the following functions:

```python
sns.regplot(data=data, x='total_bill', y='tip')
plt.show()
```

##### Facet Grid: 
- sns.FacetGrid() can be used to plot multiple subplots based on categories in your data.

```python
g = sns.FacetGrid(data, col='time')
g.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

##### Seaborns Color Palettes
- Seaborn provides various color palettes to help you create more aesthetically pleasing plots.

```python
sns.set_palette("Set2")
sns.scatterplot(data=data, x='total_bill', y='tip')
plt.show()
```

#### Seaborn Themes

- You can set themes globally for your plots:

```python
sns.set_theme(style="whitegrid", palette="muted")
sns.histplot(data['total_bill'], kde=True)
plt.show()
```
