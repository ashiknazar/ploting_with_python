###  Common Plots in Seaborn

###  Distribution Plots

#### Histogram: sns.histplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a simple dataset
data = sns.load_dataset('tips')

# Plot a histogram
sns.histplot(data['total_bill'], kde=True)
plt.show()
```

#### KDE Plot (Kernel Density Estimate)

```python
sns.kdeplot(data['total_bill'])
plt.show()
```

####  Scatter Plots

```python
sns.scatterplot(data=data, x='total_bill', y='tip', hue='time')
plt.show()
```

#### Box Plots

```python
sns.boxplot(data=data, x='time', y='total_bill')
plt.show()
```

####  Bar Plots

```python
sns.barplot(data=data, x='day', y='total_bill', estimator='mean')
plt.show()
```

- In a Seaborn bar plot (sns.barplot), if you see a line drawn at the top of each bin (or bar), it usually represents the confidence interval of the estimate for that bin.

- By default, Seaborn adds a 95% confidence interval around the mean value of each bar to indicate the uncertainty in the estimate. This line spans from the lower to the upper bound of the confidence interval. The exact nature of the line can vary depending on the type of data and the settings in the plot:

#### heatmap 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset from Seaborn
flights = sns.load_dataset("flights")

# Pivot the data to create a matrix format for the heatmap
pivot_flights = flights.pivot_table(index="month", columns="year", values="passengers")

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_flights, annot=True, cmap="YlGnBu", linewidths=0.5)

# Show the plot
plt.title("Heatmap of Passengers per Month and Year")
plt.show()
```
